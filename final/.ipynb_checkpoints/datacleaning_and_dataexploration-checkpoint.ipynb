```python
import os, json
import pandas as pd

# this finds our json files
path_to_json = 'data/raw_data/movie_reviews/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

# here I define my pandas Dataframe with the columns I want to get from the json
jsons_data = pd.DataFrame(columns=['copyright', 'has_more', 'num_results','results','status'])

# we need both the json and an index number so use enumerate()
for index, js in enumerate(json_files):
    with open(os.path.join(path_to_json, js)) as json_file:
        json_text = json.load(json_file)

        # here you need to know the layout of your json and each json has to have
        # the same structure (obviously not the structure I have here)
        copyright = json_text['features'][0]['properties']['copyright']
        has_more = json_text['features'][0]['properties']['has_more']
        num_results = json_text['features'][0]['properties']['num_results']
        results = json_text['features'][0]['properties']['results']
        status = json_text['features'][0]['properties']['status']
        # here I push a list of data into a pandas DataFrame at row given by 'index'
        jsons_data.loc[index] = [copyright, has_more, num_results,results,status]

# now that we have the pertinent json data in our DataFrame let's look at it
print(jsons_data)
```

```python
import csv
# Read file from hierarchy floder movie reviews
def find_files_path(file_root):
    path_list = []
    for root,dirs,files in os.walk(file_root):
        for filepath in files:
            if filepath.endswith('.json'):    # Some files are not json file in the data set, check file type first.
                path_list.append(os.path.join(root,filepath)) 
    return path_list
# Read each moview reviews json and get useful information
def read_moviereviews_files(file_name):
    open_hour_list = []
    with open (file_name, encoding='utf-8', errors="ignore") as file:  
        content = json.load(file)      
    return content
# Generate csv file.
def create_csv_file(list,file):
    with open(file,'w') as f:
        w = csv.writer(f)
        w.writerow(['copyright','has_more','num_results','results','status'])
        for cell in list:
            w.writerows(cell)
# Parse the arguments get final list of restaurants open hours.
whole_list = []
moviereviews_path = glob('data/raw_data/movie_reviews')
for jfile in find_files_path(moviereviews_path[0]):
    whole_list.append(read_moviereviews_files(jfile))
# Create csv file.
create_csv_file(whole_list, 'moviereviews.csv')
```
